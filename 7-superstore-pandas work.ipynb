```python
df.isnull().sum() #to find missing values
```

```python
df["Segment"].isnull().sum()
```

```python
df["Segment"].mode()
```

```python
df['Segment']=df['Segment'].fillna(df['Segment'].mode()[0])
df['Segment']
```

```python
df['Segment']=df['Segment'].fillna(df['Segment'].mode()[0])
df['Segment']
```

```python
df.isnull().sum()
```

```python
df["State"].fillna(df["State"].mode())
```

```python
df["State"].fillna(df["State"].mode(),inplace=True)
```

```python
df["State"]=df["State"].fillna("unknown")
```

```python
df["State"]
```

```python
df["Sales"]=df["Sales"].fillna(0)
```

```python
df["Sales"]
```

```python
df["Postal Code"]=df["Postal Code"].fillna(df["Postal Code"].mode()[0])
df["Postal Code"]
```

```python
median_q=df["Quantity"].median()
df["Quantity"]=df["Quantity"].fillna(median_q)
```

```python
df["Quantity"]
```

```python
df["Profit"]=df["Profit"].fillna(df["Profit"].median())
df["Profit"]
```

```python
df[["Segment","Quantity","State","Profit","Sales", "Postal Code"]].isnull().sum()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Order Date']
```

```python
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df['Ship Date']
```

```python
Cleaned_Name=df.replace(r'[^a-zA-Z0-9.]',' ',regex= True)
Cleaned_Name
```

bar chart

```python
plt.figure(figsize=(5,3))
sns.barplot(data=df,x='Category',y='Profit') # orient = h means horizontal
plt.title("Bar chart showing category over profit")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
avg=df.groupby('Category')['Profit'].mean()
avg
```

```python
plt.figure(figsize=(5,3))
avg.plot(kind='bar',color='skyblue',edgecolor='black')
plt.title("Bar chart showing profit over category")
plt.xlabel("category")
plt.ylabel("profit")
plt.show()
```

```python
plt.figure(figsize=(5,3))
avg.plot(kind='barh',color='skyblue',edgecolor='black')
plt.title("Bar chart showing profit over category")
plt.xlabel("category")
plt.ylabel("profit")
plt.show()
```

scatterplot

```python
sns.scatterplot(data=df,x='Discount',y='Profit')
plt.title("Scatter plot showing relationship between discount and profit")
plt.xlabel("discount")
plt.ylabel("Profit")
plt.show()
```

```python
sns.scatterplot(data=df,x='Discount',y='Profit',hue='Category')
plt.title("Scatter plot showing relationship between discount and profit")
plt.xlabel("discount")
plt.ylabel("Profit")
plt.show()
```

```python
sns.boxplot(data=df,x='Category',y='Profit')
plt.title("Scatter plot showing relationship between  category and profit")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```
