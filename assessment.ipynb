```python
Q1, Q3 = df['Sales'].quantile([0.25, 0.75])
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df['Sales Outlier'] = (df['Sales'] < lower) | (df['Sales'] > upper)
```

```python
df['Sales Outlier']
```

```python
from scipy import stats
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, LabelEncoder
```

```python
# 72 Profit outliers (Z-score)
df['Profit Z'] = stats.zscore(df['Profit'])
df['Profit Outlier'] = df['Profit Z'].abs() > 3
df['Profit Outlier'] 
```

```python
# 73 Remove Sales outliers
df_no_sales_outliers = df[~df['Sales Outlier']]
df_no_sales_outliers
```

```python

# 74 Remove Profit outliers
df_no_profit_outliers = df[df['Profit Outlier'] == False]
df_no_profit_outliers
```

```python
# 75 Cap Sales at 95th percentile
cap_95 = df['Sales'].quantile(0.95)
df['Sales'] = np.where(df['Sales'] > cap_95, cap_95, df['Sales'])
df['Sales']
```

```python
# 76 Cap Profit at 5th/95th percentile
low, high = df['Profit'].quantile([0.05, 0.95])
df['Profit'] = np.clip(df['Profit'], low, high)
df['Profit']
```

```python
# 77 Replace discount outliers
df['Discount'] = df['Discount'].clip(df['Discount'].quantile(0.01),df['Discount'].quantile(0.99))
df['Discount']
```

```python
# 78 Replace quantity outliers with mean
df['Quantity'] = df['Quantity'].clip(df['Quantity'].quantile(0.01),
                                     df['Quantity'].quantile(0.99))
df['Quantity']
```
